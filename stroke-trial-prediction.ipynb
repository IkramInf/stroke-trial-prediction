```python
tr_processed_data = train_data_drop_obj.drop(['ONDRUG','DIED'], axis=1)
```

```python
X = tr_processed_data.drop(['OCCODE'], axis=1)
y = tr_processed_data['OCCODE']
```

```python
X.shape
```

```python
y.shape
```

```python
test_data_drop_obj.shape
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_r = RandomForestClassifier(n_estimators=150, criterion='entropy', max_depth=130,
                               random_state=1)
score = cross_val_score(clf_r, X, y, cv=k_fold, n_jobs=1, scoring='accuracy')
print(score)
round(np.mean(score)*100, 2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=2, random_state=1)
score = cross_val_score(model, X, y, cv=k_fold, n_jobs=1, scoring='accuracy')
print(score)
print(np.mean(score))
```

```python
clf_r.fit(X,y)
```

```python
pred = clf_r.predict(test_data_drop_obj)
```

```python
np.unique(pred)
```

```python
patient_died = []
for i in pred:
    if i==1:
        patient_died.append('Y')
    else:
        patient_died.append('N')
```

```python
patient_died
```

```python
id = test['ID']

submission = pd.DataFrame({"ID" : id, 'PatientDied': patient_died})
submission.to_csv('stroke-trial-prediction.csv', index=False)
```
